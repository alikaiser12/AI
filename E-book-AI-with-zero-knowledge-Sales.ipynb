```python
import pandas as pd


week1 = pd.read_csv('sales_week1.csv') # সপ্তাহ ১ এর বিক্রি
week2 = pd.read_csv('sales_week2.csv') # সপ্তাহ ২ এর বিক্রি


# দুই সপ্তাহকে একসাথে জোড়া
sales = pd.concat([week1, week2], ignore_index=True)


# দোকানভিত্তিক সারাংশ: কতটি ট্রানজ্যাকশন (count), মোট (sum), গড় (mean)
summary = (sales
.groupby('store')['amount']
.agg(['count','sum','mean'])
.reset_index())


# ফলাফল দেখা
summary


# রিপোর্ট ফাইলে সংরক্ষণ
summary.to_csv('sales_summary.csv', index=False)
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50, 60])

print(arr[0:3])   # প্রথম ৩টি উপাদান
print(arr[2:])    # ২ নম্বর ইন্ডেক্স থেকে শেষ পর্যন্ত
print(arr[-3:])   # শেষের ৩টি উপাদান
```

বাংলা ব্যাখ্যা:

arr[0:3] মানে ইন্ডেক্স ০ থেকে ২ পর্যন্ত মান।

arr[-3:] মানে পেছন থেকে ৩টা মান।

এই slicing Excel-এর “Filter” এর মতোই ডেটা বাছাই করার উপায়।

```python
matrix = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

print(matrix[0:2, 1:3])   # প্রথম ২ সারি, কলাম ১ ও ২
print(matrix[:, 0])       # সব সারির প্রথম কলাম
print(matrix[1, :])       # দ্বিতীয় সারি সম্পূর্ণ
```

বাংলা ব্যাখ্যা:

: মানে “সব”,

[0:2, 1:3] মানে সারি ০–১ আর কলাম ১–২।

এটা টেবিলের নির্দিষ্ট অংশ কেটে নেওয়ার মতো।

```python
arr = np.array([10, 20, 30])
print(arr + 5)    # প্রতিটির সাথে ৫ যোগ
print(arr * 2)    # প্রতিটির সাথে ২ গুণ
```

বাংলা ব্যাখ্যা:

প্রতিটি মানের সাথে স্বয়ংক্রিয়ভাবে কাজ করে।

Excel-এ যেভাবে একটি সেলে ফর্মুলা কপি করো, NumPy সেটা নিজে থেকেই করে ফেলে।

```python
arr = np.array([1, 2, 3, 4, 5, 6])
reshaped = arr.reshape(2, 3)
print(reshaped)
```

```python
import numpy as np

# প্রতিটি ছাত্রের তিনটি বিষয়ের মার্কস (মোট ৬ ছাত্র × ৩ বিষয়)
marks = np.array([78, 82, 80, 85, 74, 79, 90, 88, 91, 65, 69, 72, 88, 92, 85, 77, 81, 79])

reshaped = marks.reshape(6, 3)
print("আকার:", reshaped.shape)
print(reshaped)
```

```python
import numpy as np

sales = np.array([100,120,130,110,115,150,160,140,155,170,180,175])
reshaped = sales.reshape(4,3)  # ৪ সপ্তাহ × ৩ দিন
print(reshaped)

weekly_avg = np.mean(reshaped, axis=1)
print("সাপ্তাহিক গড় বিক্রি:", weekly_avg)
```
